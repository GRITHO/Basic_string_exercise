# A.donuts

Given an int count of a number of donuts, return a string of the form 'Number of donuts: ', where  is the number passed in

```python
def donuts(count):
    if count < 10:
        return 'Number of donuts: {}'.format(count)
    else:
        return 'Number of donuts: many'
```

```python
donuts(5)
```

```python
donuts(30)
```

# B.both_ends

Given a string s, return a string made of the first 2 and the last 2 chars of the original string

```python
my_string = 'Thomas_Griesser'
```

```python
def both_ends(s):
    return s[:2] + s[-2:] if len(s) >= 2 else ''
```

```python
both_ends(my_string)
```

# C.fix_start

Given a string s, return a string where all occurences of its first char have been changed to '*', except do not change the first char itself

```python
def fix_start(s):
    fl = s[0]
    return fl + s[1:].replace(fl, "*")
```

```python
fix_start(my_string)
```

```python
s_string = 'All_NAme'
```

```python
fix_start(s_string)
```

# D.MixUp

Given strings a and b, return a single string with a and b separated by a space, except swap the first 2 chars of each string.

```python
def mix_up(a, b):
    out_a = b[:2] + a[2:]
    out_b = a[:2] + b[2:]
    return "{} {}".format(out_a, out_b)
```

```python
mix_up(my_string, s_string)
```

# D.verbing

Given a string, if its length is at least 3, add 'ing' to its end. Unless it already ends in 'ing', in which case add 'ly' instead. If the string length is less than 3, leave it unchanged

```python
def verbing(s):
    if len(s) >= 3:
        if s[-3:] != 'ing':
            s += 'ing'  
        else:
            s += 'ly'
    return s
```

```python
verbing(my_string)
```

```python
verbing(s_string)
```

```python
verbing('beging')
```

# E.not_bad

Given a string, find the first appearance of the substring 'not' and 'bad'. If the 'bad' follows the 'not', replace the whole 'not'...'bad' substring with 'good'.

Define a sentence

```python
my_sent = 'This table is not bad'
```

```python
def not_bad(s):
    w_not = s.find('not')
    w_bad = s.find('bad')
    if w_not != -1 and w_bad != -1 and w_bad > w_not:
        s = s[:w_not] + 'good' + s[w_bad+3:]
    return s
```

```python
not_bad(my_sent)
```

# F.front_back

Consider dividing a string into two halves. If the length is even, the front and back halves are the same length. If the length is odd, we'll say that the extra char goes in the front half

Define two strings

```python
fst_st = 'lunch'
```

```python
sd_st = 'dinner'
```

```python
import math
```

```python
def front_back(a, b):
    c_a = math.ceil(len(a)/2)
    c_b = math.ceil(len(b)/2)
    return '{}{}{}{}'.format(a[:c_a], b[:c_b], a[c_a:], b[c_b:])
```

```python
front_back(fst_st, sd_st)
```
